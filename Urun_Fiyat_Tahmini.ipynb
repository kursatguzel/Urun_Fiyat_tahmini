```python
X=df_train[['ürün','ürün kategorisi','ürün üretim yeri','market','şehir']]
y=df_train['ürün fiyatı']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['ürün','ürün kategorisi','ürün üretim yeri','market','şehir']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['ürün','ürün kategorisi','ürün üretim yeri','market','şehir']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
